```python
import numpy as np
import pandas as pd
import requests # 크롤링에 사용하는 패키지
from bs4 import BeautifulSoup # html 변환에 사용함
from selenium import webdriver
from selenium.webdriver.common.keys import Keys #page_up, page_down
import time
```

```python
df_n = pd.read_csv('./netflix_unique.csv')
df_t = pd.read_csv('./tving_unique.csv')
df_tn = pd.read_csv('./t_n_uniq.csv')
```

```python
df_tn
```

```python
sns_list = df_tn['t_n_unique'].tolist()
```

```python
driver = webdriver.Chrome('.//chromedriver.exe')
url = 'https://naver.com'
driver.get(url)
response = requests.get(url)
```

```python
like = []
for title in sns_list:
    print(f'{title} 크롤링 하는 중')
    key_word = title
    driver.find_element_by_xpath('//*[@id="query"]').send_keys(key_word +'\n')
    time.sleep(3)
    driver.find_element_by_xpath('//*[@id="query"]')
    html = BeautifulSoup(driver.page_source, 'html.parser')
    
    # 찜정보에 대한 정보가 없는 경우 number에 0 기입
    if not html.select('span.api_spim_af'):
        number = 0
        like.append(number)
    # 찜정보에 대한 정보가 있는 경우 number에 해당 숫자 기입
    else:
        for count in html.select('span.api_spim_af'):
            number = int(count.text.replace(',',''))
            like.append(number)
    print(f'{number}개')
    driver.find_element_by_xpath('//*[@id="header_wrap"]/div/div[1]/div[1]/div/h1/a/i').click()
    
```

```python
df_tn['naver_like'] = like
```

```python
df_tn
```

```python
for i in range(0,112):
    if df_tn['naver_like'][i] == 0:
        key_word = df_tn['t_n_unique'][i]
        driver.find_element_by_xpath('//*[@id="query"]').send_keys('영화' + key_word +'\n')
        time.sleep(3)
        driver.find_element_by_xpath('//*[@id="query"]')
        html = BeautifulSoup(driver.page_source, 'html.parser')
    
    # 찜정보에 대한 정보가 없는 경우 number에 0 기입
        if not html.select('span.api_spim_af'):
            number = 0
            df_tn['naver_like'][i] = number
    # 찜정보에 대한 정보가 있는 경우 number에 해당 숫자 기입
        else:
            for count in html.select('span.api_spim_af'):
                number = int(count.text.replace(',',''))
                df_tn['naver_like'][i] = number
        print(f'{number}개')
        driver.find_element_by_xpath('//*[@id="header_wrap"]/div/div[1]/div[1]/div/h1/a/i').click()
```
