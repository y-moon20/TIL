```python
import numpy as np
import pandas as pd
import requests # 크롤링에 사용하는 패키지
from bs4 import BeautifulSoup # html 변환에 사용함
from selenium import webdriver
from selenium.webdriver.common.keys import Keys #page_up, page_down
import time
```

```python
driver = webdriver.Chrome('.//chromedriver.exe')
```

```python
url = 'https://m.kinolights.com/themes/18'
driver.get(url)
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
```

```python
#37주 페이지 들어갔다 나오기
for i in range(37, 0, -1):
    driver.find_element_by_xpath(f'//*[@id="contents"]/div[2]/div/div[{i}]/a/div[2]/div/div/span').click()
    time.sleep(3)
    driver.find_element_by_xpath('//*[@id="header"]/div/button[1]/i').click()
    time.sleep(3)
```

```python
# 넷플릭스 1주일분 크롤링
driver.find_element_by_xpath('//*[@id="triggerMenu"]/div/div[1]/div/div/button[1]/i').click()
time.sleep(3)
html = BeautifulSoup(driver.page_source, 'html.parser')
Netflix = []
title_list = []
for title in html.select('div.title'):
    title_list.append(title.text.strip())
Netflix.append(title_list)
driver.find_element_by_xpath('//*[@id="triggerMenu"]/div/div[1]/div/div/button[1]/i').click()
Netflix
```

```python
#티빙 1주일분 크롤링
driver.find_element_by_xpath('//*[@id="triggerMenu"]/div/div[1]/div/div/button[4]/i').click()
time.sleep(3)
html = BeautifulSoup(driver.page_source, 'html.parser')
Tiving = []
title_list = []
for title in html.select('div.title'):
    title_list.append(title.text.strip())
Tiving.append(title_list)
driver.find_element_by_xpath('//*[@id="triggerMenu"]/div/div[1]/div/div/button[4]/i').click()    
Tiving
```

```python
Netflix = []
Tiving = []
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

for i in range(37, 0, -1):
    if i%13 == 0:
        driver.find_element_by_tag_name('body').send_keys(Keys.PAGE_UP)
        
    #각 주차별 랭킹페이지로 진입    
    driver.find_element_by_xpath(f'//*[@id="contents"]/div[2]/div/div[{i}]/a/div[2]/div/div/span').click()
    time.sleep(3)
    
    #Netflix버튼 활성화
    driver.find_element_by_xpath('//*[@id="triggerMenu"]/div/div[1]/div/div/button[1]/i').click()
    print(f'{i}주 전 Netflix Data를 크롤링 중 입니다.')
    time.sleep(3)
    html = BeautifulSoup(driver.page_source, 'html.parser')
    
    #Netflix데이터 저장
    title_list = []
    for title in html.select('div.title'):
        title_list.append(title.text.strip())
    Netflix.append(title_list)
    
    #Netflix버튼 비활성화
    driver.find_element_by_xpath('//*[@id="triggerMenu"]/div/div[1]/div/div/button[1]/i').click()
    time.sleep(3)
    
    #Tiving버튼 활성화
    driver.find_element_by_xpath('//*[@id="triggerMenu"]/div/div[1]/div/div/button[4]/i').click()
    time.sleep(3)
    print(f'{i}주 전 Tiving Data를 크롤링 중 입니다.')
    html = BeautifulSoup(driver.page_source, 'html.parser')
    
    #Tiving 데이터 저장
    title_list = []
    for title in html.select('div.title'):
        title_list.append(title.text.strip())
    Tiving.append(title_list)
    
    #Tiving버튼 비활성화
    driver.find_element_by_xpath('//*[@id="triggerMenu"]/div/div[1]/div/div/button[4]/i').click()
    time.sleep(3) 
    
    #메인 랭킹페이지로 돌아가기
    driver.find_element_by_xpath('//*[@id="header"]/div/button[1]/i').click()
    time.sleep(3)
```

```python
#중복제거한 Netflix 인기 콘텐츠리스트
Netflix_title = []
for title in Netflix:
    for i in title:
        if i not in Netflix_title:
            Netflix_title.append(i)
```

```python
Netflix_title
```

```python
#중복제거한 Tiving 인기 콘텐츠리스트
Tiving_title = []
for title in Tiving:
    for i in title:
        if i not in Tiving_title:
            Tiving_title.append(i)
```

```python
Tiving_title
```

```python
# Netflix의 인기콘텐츠가 Netflix에만 존재하는지 확인
only_netflix = []
for n in Netflix_title:
    if n not in Tiving_title:
        only_netflix.append('1')
    else:
        only_netflix.append('0')
```

```python
only_netflix
```

```python
# Tiving의 인기콘텐츠가 Tiving에만 존재하는지 확인
only_netflix = []
only_tiving = []
for t in Tiving_title:
    if t not in Netflix_title:
        only_tiving.append('1')
    else:
        only_tiving.append('0')
```

```python
only_tiving
```

```python
#Netflix dataframe
count = list(range(1,len(Netflix_title)+1))
columns = ['title']
df_N = pd.DataFrame(Netflix_title,
                 index = count,
                 columns = columns)
df_N['original'] = only_netflix
df_N
```

```python
#Tiving dataframe
count = list(range(1,len(Tiving_title)+1))
columns = ['title']
df_T = pd.DataFrame(Tiving_title,
                 index = count,
                 columns = columns)
df_T['original'] = only_tiving
df_T
```

```python
#Netflix와 Tiving 리스트의 중복을 제거한, sns에 검색할 이름
sns_title = []
for title in Netflix_title:
    if title not in sns_title:
        sns_title.append(title)
for title in Tiving_title:
    if title not in sns_title:
        sns_title.append(title)
```
